```python
import pandas as pd
import numpy as np

#!pip install plotly
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.offline as ply
```

```python
# Read the clean file created from import
noise_trim = pd.read_csv(r'C:\Users\Owner\Desktop\Ys Take Home\noise_trim.csv')
```

```python
# Plot for the number of cases by complaint type
complaint_type_x = list(noise_trim["Complaint_Type"].value_counts())
complaint_type_y = list(noise_trim["Complaint_Type"].value_counts().index)

complaint_type_x.reverse()
complaint_type_y.reverse()

complaint_type_count_df = pd.DataFrame({"Total Calls": complaint_type_x, "Complaint Type": complaint_type_y})

complaint_type_fig = px.bar(complaint_type_count_df, x="Total Calls", y="Complaint Type", 
             orientation='h',
             height=400, color = "Total Calls",
             color_discrete_sequence = px.colors.sequential.thermal)



complaint_type_fig.show()
```

```python
# Plot for the nmber of cases by complaint description
complaint_desc_x = list(noise_trim["Descriptor"].value_counts())
complaint_desc_y = list(noise_trim["Descriptor"].value_counts().index)

complaint_desc_x.reverse()
complaint_desc_y.reverse()

complaint_type_count_df = pd.DataFrame({"Total Calls": complaint_desc_x, "Complaint Descriptor": complaint_desc_y})

complaint_type_fig = px.bar(complaint_type_count_df, x="Total Calls", y="Complaint Descriptor", 
             #orientation='h',
             height=900, color = "Total Calls",
             color_discrete_sequence = px.colors.sequential.speed)

complaint_type_fig.update_layout(hovermode='x',
                  title="Top 5 Most Frequent Complaint Descriptors",
                  font=dict(family="silom",
                  size=14, color="black"))

complaint_type_fig.show()
```

```python
#!pip install sqldf 
import sqldf
```

```python
# Plot the top 5 drivers using the Complaint_Type field

qtop5 = """
select Complaint_Type, count(Unique_Key) as Total_Calls from noise_trim
group by Complaint_Type
order by Total_Calls DESC
limit 5
"""
top5 = sqldf.run(qtop5)

qarea = """
select Created_Date, noise_trim.Complaint_Type, count(Unique_Key) as Total_Calls
from noise_trim 
join top5 
on noise_trim.Complaint_Type = top5.Complaint_Type
group by Created_Date, noise_trim.Complaint_Type
"""
areadf = sqldf.run(qarea)

fig = px.area(areadf, x="Created_Date", y= "Total_Calls")

fig.show()
```

```python
# Plot the top 5 drivers of noise using Descriptors

qtop5 = """
select Descriptor, count(Unique_Key) as Total_Calls from noise_trim
group by Descriptor
order by Total_Calls DESC
limit 5
"""
top5 = sqldf.run(qtop5)

qarea = """
select Created_Date, noise_trim.Descriptor, count(Unique_Key) as Total_Calls
from noise_trim 
join top5 
on noise_trim.Descriptor = top5.Descriptor
group by Created_Date, noise_trim.Descriptor
"""
areadf = sqldf.run(qarea)

fig = px.line(areadf, x="Created_Date", y= "Total_Calls", color = "Descriptor")

fig.show()
```
